```python
from pybaseball import batting_stats
import pandas as pd

data = batting_stats(2020, 2022, ind=1, qual=100)
```

```python
hrPerPA = []
for row in data.index:
    val = (data['HR'][row] / data['PA'][row])
    hrPerPA.append(val)
data['HR/PA'] = hrPerPA
```

```python
# Taking the average of all projected stats in 2021 in order to regress to the mean

AVG_BA = data[data['Season'] == 2022]['AVG'].mean()

AVG_SLG = data[data['Season'] == 2022]['SLG'].mean()

AVG_OBP = data[data['Season'] == 2022]['OBP'].mean()

AVG_hrPerPA = data[data['Season'] == 2022]['HR/PA'].mean()

AVG_Kp = data[data['Season'] == 2022]['K%'].mean()

AVG_BBp = data[data['Season'] == 2022]['K%'].mean()
```

```python
# Creates projections dataframe
proj_df = pd.DataFrame(columns={'Name': [], 'proj PA':[], '3yrPA':[], '2022 AVG': [], '2021 AVG': [], '2020 AVG': [], 
                                'proj AVG': [], '2022 SLG': [], '2021 SLG': [], '2020 SLG': [], 'proj SLG': [], 
                                '2022 OBP': [], '2021 OBP': [], '2020 OBP': [], 'proj OBP': [], '2022 HR/PA':[], 
                                '2021 HR/PA':[], '2020 HR/PA':[], 'proj HR/PA':[], 
                                '2022 K%':[], '2021 K%':[], '2020 K%':[], 'proj K%':[], '2022 BB%':[], '2021 BB%':[], 
                                '2020 BB%':[], 'proj BB%':[]})

# Only projects players with 2021 data, Marcel includes all players but I thought it 
# would be easier to work with
players_2022 = []
for k in range(len(data.Name)):
    if data['Season'][k] == 2022:
        players_2022.append(data['Name'][k])
```

```python
proj_df['Name'] = players_2022
```

```python
df = data[data['Name'].isin(proj_df.Name)]
```

```python
# Loop through the players in the df which contains only the data from players
# that played in 2021
index = 0
for name in proj_df.Name:
    # Count the number of hits for the player in 2021
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2022]
    avg = age_df['AVG']
    proj_df['2022 AVG'][index] = avg.item()
    index += 1
    
index = 0
for name in proj_df.Name:
    # Count the number of hits for the player in 2020
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2021]
    avg = age_df['AVG'].sum()
    proj_df.loc[index, '2021 AVG'] = avg
    index += 1
    
index = 0
for name in proj_df.Name:
    # Count the number of hits for the player in 2019
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2020]
    avg = age_df['AVG'].sum()
    proj_df.loc[index, '2020 AVG'] = avg
    index += 1
```

```python
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2022]
    obp = age_df['OBP']
    proj_df['2022 OBP'][index] = obp.item()
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2021]
    obp = age_df['OBP'].sum()
    proj_df.loc[index, '2021 OBP'] = obp
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2020]
    obp = age_df['OBP'].sum()
    proj_df.loc[index, '2020 OBP'] = obp
    index += 1
```

```python
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2022]
    slg = age_df['SLG']
    proj_df['2022 SLG'][index] = slg.item()
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2021]
    slg = age_df['SLG'].sum()
    proj_df.loc[index, '2021 SLG'] = slg
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2020]
    slg = age_df['SLG'].sum()
    proj_df.loc[index, '2020 SLG'] = slg
    index += 1
```

```python
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2022]
    hrPerPA = age_df['HR/PA']
    proj_df['2022 HR/PA'][index] = hrPerPA.item()
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2021]
    hrPerPA = age_df['HR/PA'].sum()
    proj_df.loc[index, '2021 HR/PA'] = hrPerPA
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2020]
    hrPerPA = age_df['HR/PA'].sum()
    proj_df.loc[index, '2020 HR/PA'] = hrPerPA
    index += 1
```

```python
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2022]
    k = age_df['K%']
    proj_df['2022 K%'][index] = k.item()
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2021]
    k = age_df['K%']
    if k.empty or k.item() <= 0:
        proj_df['2021 K%'][index] = 0
    else:
        proj_df['2021 K%'][index] = k.item()
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2020]
    k = age_df['K%']
    if k.empty or k.item() <= 0:
        proj_df['2020 K%'][index] = 0
    else:
        proj_df['2020 K%'][index] = k.item()
    index += 1
```

```python
# Loop through the players in the DataFrame
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2022]
    bb = age_df['BB%']
    proj_df['2022 BB%'][index] = bb.item()
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2021]
    bb = age_df['BB%']
    if bb.empty or bb.item() <= 0:
        proj_df['2021 BB%'][index] = 0
    else:
        proj_df['2021 BB%'][index] = bb.item()
    index += 1
    
index = 0
for name in proj_df.Name:
    name_df = df[df['Name'] == name]
    age_df = name_df[name_df['Season'] == 2020]
    bb = age_df['BB%']
    if bb.empty or bb.item() <= 0:
        proj_df['2020 BB%'][index] = 0
    else:
        proj_df['2020 BB%'][index] = bb.item()
    index += 1
```

```python
# if you don't have enough data (1200 pa), add the difference between abs and 1200 times average BA
for row in proj_df.index:
    name_df = name_df = data[data['Name'] == data.Name[row]].reset_index(drop=True)
    pa = name_df['PA'].sum()
    proj_df['3yrPA'][row] = pa
```

```python
for row in proj_df.index:
    weight = 0
    avg_2022 = proj_df['2022 AVG'][row]
    if avg_2022 > 0:
        weight += 5
    avg_2021 = proj_df['2021 AVG'][row]
    if avg_2021 > 0: 
        weight += 4
    avg_2020 = proj_df['2020 AVG'][row]
    if avg_2020 > 0:
        weight += 3
    proj_df['proj AVG'][row] = (5*avg_2022 + 4*avg_2021 + 3*avg_2020) / weight
```

```python
for row in proj_df.index:
    weight = 0
    obp_2022 = proj_df['2022 OBP'][row]
    if obp_2022 > 0:
        weight += 5
    obp_2021 = proj_df['2021 OBP'][row]
    if obp_2021 > 0: 
        weight += 4
    obp_2020 = proj_df['2020 OBP'][row]
    if obp_2020 > 0:
        weight += 3
    proj_df['proj OBP'][row] = (5*obp_2022 + 4*obp_2021 + 3*obp_2020) / weight
```

```python
for row in proj_df.index:
    weight = 0
    slg_2022 = proj_df['2022 SLG'][row]
    if slg_2022 > 0:
        weight += 5
    slg_2021 = proj_df['2021 SLG'][row]
    if slg_2021 > 0: 
        weight += 4
    slg_2020 = proj_df['2020 SLG'][row]
    if slg_2020 > 0:
        weight += 3
    proj_df['proj SLG'][row] = (5*slg_2022 + 4*slg_2021 + 3*slg_2020) / weight
```

```python
for row in proj_df.index:
    weight = 0
    hrPerPA_2022 = proj_df['2022 HR/PA'][row]
    if hrPerPA_2022 > 0:
        weight += 5
    hrPerPA_2021 = proj_df['2021 HR/PA'][row]
    if hrPerPA_2021 > 0: 
        weight += 4
    hrPerPA_2020 = proj_df['2020 HR/PA'][row]
    if hrPerPA_2020 > 0:
        weight += 3
    proj_df['proj HR/PA'][row] = (5*hrPerPA_2022 + 4*hrPerPA_2021 + 3*hrPerPA_2020) / weight
```

```python
for row in proj_df.index:
    weight = 0
    k_2022 = proj_df['2022 K%'][row]
    if k_2022 > 0:
        weight += 5
    k_2021 = proj_df['2021 K%'][row]
    if k_2021 > 0:
        weight += 4
    k_2020 = proj_df['2020 K%'][row]
    if k_2020 > 0:
        weight += 3
    proj_df['proj K%'][row] = (5*k_2022 + 4*k_2021 + 3*k_2020) / weight
```

```python
for row in proj_df.index:
    weight = 0
    bb_2022 = proj_df['2022 BB%'][row]
    if bb_2022 > 0:
        weight += 5
    bb_2021 = proj_df['2021 BB%'][row]
    if bb_2021 > 0:
        weight += 4
    bb_2020 = proj_df['2020 BB%'][row]
    if bb_2020 > 0:
        weight += 3
    proj_df['proj BB%'][row] = (5*bb_2022 + 4*bb_2021 + 3*bb_2020) / weight
```

```python
for row in proj_df.index: 
    if proj_df['3yrPA'][row] < 1200:
        proj_df['proj AVG'][row] += (AVG_BA/(1200-proj_df['3yrPA'][row]))
        proj_df['proj SLG'][row] += (AVG_SLG/(1200-proj_df['3yrPA'][row]))
        proj_df['proj OBP'][row] += (AVG_OBP/(1200-proj_df['3yrPA'][row]))
        proj_df['proj HR/PA'][row] += (AVG_hrPerPA/(1200-proj_df['3yrPA'][row]))
        proj_df['proj K%'][row] += (AVG_Kp/(1200-proj_df['3yrPA'][row]))
        proj_df['proj BB%'][row] += (AVG_BBp/(1200-proj_df['3yrPA'][row]))
```

```python
# Creates a name (key) and age (value) map for purposes of age adjustment
name_to_age = {}
for name in proj_df['Name']:
    name_df = df[df['Name'] == name]
    for row in name_df.index:
        if name_df['Season'][row] == 2022:
            age = name_df['Age'][row]
    name_to_age[name] = age
```

```python
for row in proj_df.index:
    name_df = df[df['Name'] == proj_df.Name[row]]
    season_22 = name_df[name_df['Season'] == 2022]
    season_21 = name_df[name_df['Season'] == 2021]
    pa_2022 = season_22['PA'].item()
    if season_21.empty:
        pa_2021 = 0
    else:
        pa_2021 = season_21['PA'].item()   
    proj_PA = (0.5 * pa_2022) + (0.1 * pa_2021) + 200
    proj_df['proj PA'][row] = round(proj_PA, 0)
```

```python
index = 0
for name in proj_df['Name']:
    if name_to_age[name] >= 29:
        proj_df['proj PA'][index] -= (name_to_age[name] - 29) * .003
        index += 1
    else:
        proj_df['proj PA'][index] -= (name_to_age[name] - 29) * .006
        index += 1
```

```python
for row in proj_df.index:
    proj_df['proj PA'][row] = round(proj_df['proj PA'][row], 0)
```

```python
import numpy as np

proj_df['proj HR'] = np.nan
proj_df['proj K'] = np.nan
proj_df['proj BB'] = np.nan

for row in proj_df.index:
    proj_df['proj HR'][row] = round(proj_df['proj PA'][row] * proj_df['proj HR/PA'][row], 0)
    proj_df['proj K'][row] = round(proj_df['proj PA'][row] * proj_df['proj K%'][row], 0)
    proj_df['proj BB'][row] = round(proj_df['proj PA'][row] * proj_df['proj BB%'][row], 0)
```
